```python
import pandas as pd
import os
import glob
```

```python
#extracting all files from local directory for csse are merging them
extension = 'csv'
path = "/Users/rayni/Desktop/COVID-19-master/csse_covid_19_data/csse_covid_19_daily_reports"
allFilesInFolder = [i for i in glob.glob((path + '*.{}').format(extension))]
```

```python
print("The csv files ending with .csv are: ", allFilesInFolder)
```

```python
combinedOfCsv = pd.concat([pd.read_csv(file).assign(date = file) for file in allFilesInFolder],ignore_index = True)
```

```python
combinedOfCsv['date'] = combinedOfCsv['date'].str.replace(r'.csv','').astype('datetime64[ns]')
```
